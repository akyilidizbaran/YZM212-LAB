## 2. `numpy.linalg.eig` ozeti

NumPy dokumanina gore `eig`, kare bir matrisin ozdegerlerini ve sag ozvektorlerini hesaplar. Ozvektorler sutun bazli dondurulur ve hesaplama LAPACK `_geev` rutinleri uzerinden gerceklesir. Kaynaklar:

- https://numpy.org/doc/2.1/reference/generated/numpy.linalg.eig.html
- https://github.com/numpy/numpy/blob/v2.1.3/numpy/linalg/_linalg.py

```python
from eigenvector_values_analysis import format_eigenvalues, run_analysis

results = run_analysis()

for result in results:
    print('=' * 72)
    print(result['name'])
    print('Manual eigenvalues:', format_eigenvalues(result['manual_eigenvalues']))
    print('NumPy eigenvalues: ', format_eigenvalues(result['numpy_eigenvalues']))
    print(f"Max abs difference: {result['max_abs_difference']:.3e}")
    print(f"Manual avg runtime: {result['manual_avg_ms']:.6f} ms")
    print(f"NumPy avg runtime:  {result['numpy_avg_ms']:.6f} ms")
    print(f"Manual / NumPy:     {result['slowdown_ratio']:.2f}x")
    print(f"Max residual norm:  {result['max_residual_norm']:.3e}")
```
